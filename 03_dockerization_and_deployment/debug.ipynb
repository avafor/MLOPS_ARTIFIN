```python
import os, mlflow
from mlflow.tracking import MlflowClient

mlflow.set_tracking_uri(os.environ["MLFLOW_TRACKING_URI"])
client = MlflowClient()

mv = client.get_latest_versions("iris_baseline", stages=["Staging"])[0]
print("Model version:", mv.version)
print("Run ID:", mv.run_id)
print("Source:", mv.source)  # <-- this is the critical path
```
